## Financial Econometrics I: Final Project

```r
# packages
library(xts)
library(zoo)
library(dplyr)
```

```r
# load data
unzip("data_project.zip", exdir = "data_project")

loaded_objects <- load("data_project/20.RData")
loaded_objects

data20 <- get(loaded_objects[1])

head(data20)
```

```r
df <- data.frame(
  date = index(data20),
  coredata(data20)
)

head(df)
```

```r
# Create variables for analysis ----
# log RV
df$logRV <- log(df$RV)

# lagged variables
df$logRV_lag1 <- dplyr::lag(df$logRV, 1)

df$logRV_week <- zoo::rollmean(dplyr::lag(df$logRV, 1), 
                               k = 5, 
                               align = "right", 
                               fill = NA)

df$logRV_month <- zoo::rollmean(dplyr::lag(df$logRV, 1), 
                                k = 22, 
                                align = "right", 
                                fill = NA)
df$RS_lag1 <- dplyr::lag(df$RS, 1)
df$RK_lag1 <- dplyr::lag(df$RK, 1)
```

```r
df <- na.omit(df)
```

```r
summary(df[, c("ret", "RV", "logRV", "RS", "RK")])
```

```r
# Visualisations ----
plot(df$date, df$ret, type = "l", main = "Returns", xlab = "Date", ylab = "ret")
plot(df$date, df$RV, type = "l", main = "Realized Volatility", xlab = "Date", ylab = "RV")
plot(df$date, df$logRV, type = "l", main = "Log Realized Volatility", xlab = "Date", ylab = "log(RV)")
plot(df$date, df$RS, type = "l", main = "Realized Skewness", xlab = "Date", ylab = "RS")
plot(df$date, df$RK, type = "l", main = "Realized Kurtosis", xlab = "Date", ylab = "RK")
```

```r
# Autocorrelation functions ----
acf(df$ret, main = "ACF of Returns")
acf(df$RV, main = "ACF of RV")
acf(df$logRV, main = "ACF of log RV")
```
